```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "./desmatamento-amazonia.csv"
```

```python
data = pd.read_csv(file_path)
```

```python
# primeiras linhas dos dados
print(data.head())
```

```python
# mostrando informações do dataframe
print("\nInformações do DataFrame:")
print(data.info())
```

```python
# estatísticas gerais sobre os dados
print("\nEstatísticas descritivas:")
print(data.describe())
```

```python
# checando se tem valores nulo
print("\nValores ausentes em cada coluna:")
print(data.isnull().sum())
```

```python
# checando nomes das colunas
print("Nomes Colunas")
print(data.columns)
```

```python
# plotando grafico de distribução da area desmatada em 2022
plt.figure(figsize=(10, 6))
sns.histplot(data['Desmatado 2022'], bins=30, kde=True)
plt.title('Distribuição da Área Desmatada em 2022')
plt.xlabel('Área Desmatada (hectares)')
plt.ylabel('Frequência')
plt.show()
```

```python
# comparação entre Municípios
plt.figure(figsize=(14, 8))
sns.barplot(x='Municipio', y='Desmatado 2022', data=data, estimator=np.sum)  # Nome da coluna ajustado
plt.xticks(rotation=90)
plt.title('Área Desmatada em Diferentes Municípios')
plt.xlabel('Municípios')
plt.ylabel('Área Desmatada (hectares)')
plt.show()
```

```python
# proporção entre Floresta e Desmatamento
data['Proporcao'] = data['Desmatado 2022'] / data['Area Km2']
plt.figure(figsize=(10, 6))
sns.histplot(data['Proporcao'], bins=30, kde=True)
plt.title('Proporção de Desmatamento em Relação à Área Total')
plt.xlabel('Proporção de Desmatamento')
plt.ylabel('Frequência')
plt.show()
```
